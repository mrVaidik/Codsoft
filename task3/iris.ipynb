```python
import pandas as pd
import numpy as np
```

Load the Dataset

```python
df = pd.read_csv("IRIS.csv")
```

```python
df
```

Checking Null Values 

```python
df.isnull().sum()
```

```python
df['species'].value_counts()
```

```python
def species_swap(a):
    if a == "Iris-setosa":
        return 0
    if a == "Iris-versicolor":
        return 1
    else:
        return 2
    
df['species'] = df['species'].apply(species_swap)
```

```python
df['species']
```

```python
sno = range(1,150)

df['sno'] = pd.Series(sno)
```

```python
df['sno']
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='sno',y='sepal_length', hue='species', palette='deep', data=df)
plt.xlabel('Count')
plt.title('Sepal length of various species of Iris flower')
plt.legend(title="Lenends", labels=['Setosa','Versicolor','Virginica'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='sno' , y = 'sepal_width' , data=df ,hue='species' , palette='deep')
plt.xlabel('Count')
plt.title('Sepal width of various species of Iris flower')
plt.legend(title="Lenends", labels=['Setosa','Versicolor','Virginica'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='sno' , y = 'petal_length' , data=df ,hue='species' , palette='deep')
plt.xlabel('Count')
plt.title('petal length of various species of Iris flower')
plt.legend(title="Lenends", labels=['Setosa','Versicolor','Virginica'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='sno' , y = 'petal_width' , data=df ,hue='species' , palette='deep')
plt.xlabel('Count')
plt.title('petal width of various species of Iris flower')
plt.legend(title="Lenends", labels=['Setosa','Versicolor','Virginica'])
plt.show()
```

Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns='species',axis=1)
y = df['species']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,shuffle=True,random_state=42)
```

```python
model = DecisionTreeClassifier()
model.fit(xtrain,ytrain)
```

```python
pred = model.predict(xtest)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
a = accuracy_score(ytest,pred)
print(a)
```
